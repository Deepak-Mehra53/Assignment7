```python
'''Q1. What is Abstraction in OOps? Explain with an example.
Ans:- Data abstraction in Python is a programming concept that hides complex implementation details while exposing only 
      essential information and functionalities to users. 
      
For Example:- '''

from abc import ABC, abstractmethod


class Shape(ABC):
    
    @abstractmethod
    def calculate_area(self):
        pass

    
    def display_area(self):
        print("Area:", self.calculate_area())


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    
    def calculate_area(self):
        return 3.14 * self.radius * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    
    def calculate_area(self):
        return self.length * self.width


rectangle = Rectangle(4, 6)

circle.display_area()     
rectangle.display_area() 
```

```python
'''Q3. What is abc module in python? Why is it used?'''
'''Ans:- abc module :-  Abstract classes are classes that contain one or more abstract methods. An abstract method is a method 
        that is declared, but contains no implementation. Abstract classes cannot be instantiated, and require subclasses 
        to provide implementations for the abstract methods.
        
        why is it used :-  
        1.abc helps create rules (abstract base classes) that other classes must follow.
          Avoiding Weird Objects:

        2.Abstract classes can't be used on their own; they need to be completed by other classes. 
          This avoids creating strange or incomplete objects.

        3.If a class says it follows certain rules (abstract methods), it must actually do so. 
          This ensures important methods are always available.
          
        4.Using abstract base classes makes your code clear and easy to understand, especially when different classes need to have
          a similar structure.'''
```

```python
'''Q5. Can we create an instance of an abstract class? Explain your answer.
Ans:-  No, we can't create an instance of an abstract class in Python.
       Imagine an abstract class as a set of instructions saying, "Hey, if we want to be a complete class, we must
       have these methods." But the abstract class itself is not fully defined; it's missing the actual code for those methods.
       
       So, trying to create an instance (an object) directly from an abstract class doesn't work because it's like trying to 
       use a recipe without all the ingredients – we need a complete class (a concrete subclass) before we can make an actual object.
```
